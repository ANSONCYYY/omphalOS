omphalOS warehouse exploration

Load a run warehouse.sqlite and execute curated queries.

```
import sqlite3, json
from pathlib import Path

run_dir = Path('runs')

db = next(run_dir.glob('**/warehouse/warehouse.sqlite'), None)
print(db)
```

```
conn = sqlite3.connect(str(db))
conn.row_factory = sqlite3.Row
rows = conn.execute('select * from v_entity_trade_rollup order by total_value_usd desc limit 10').fetchall()
for r in rows:
    print(dict(r))
conn.close()
```
